```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1+np.exp (-x))
```

```python
X = np.linspace(-10,10)
```

```python
plt.plot(X, sigmoid(X))
```

```python
plt.axis('tight')
```

```python
plt.title('Activation Function :Sigmoid')
```

```python
plt.show()
```
